```python
## Defining X and Y 
X = main_data.drop("class",axis=1)
Y = main_data["class"]
```

```python
## Train - Test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=1)
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
classifier = LogisticRegression(solver="liblinear",random_state=1)
classifier.fit(X_train,Y_train)
```

```python
accuracies =cross_val_score(estimator=classifier,X=X_train,y=Y_train,cv=12)
print("Accuracies:\n",accuracies)
Y_pred_logi=classifier.predict(X_test)

print("Mean Accuracy",accuracies.mean())
```

```python
print("Accuracy score of LogisticRegression ")
accuracy_score(Y_test,Y_pred_logi)
```

```python
from sklearn.metrics import confusion_matrix
Y_pred_logi=classifier.predict(X_test)
confusion_matrix(Y_test,Y_pred_logi)
```

```python
print("Confusion matrix for Logistic Regression")
cm=confusion_matrix(Y_test,Y_pred_logi)
sns.heatmap(cm,annot=True)
```

## Support Vector Machine

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel="linear")
svm_classifier.fit(X_train,Y_train)

svm_accuracies = cross_val_score(estimator=svm_classifier,X=X_train,y=Y_train,cv=12)
print("Accuracies :\n",svm_accuracies)
```

```python
svc_pred=svm_classifier.predict(X_test)
print("Mean accuracy:",svm_accuracies.mean())
```

```python
print("Accuracy score of Support Vector Machine ")
svc_pred=svm_classifier.predict(X_test)
confusion_matrix(Y_test,svc_pred)
```

```python
print("Confusion metrix of SVM")
cm=confusion_matrix(Y_test,svc_pred,labels=[0,1])
sns.heatmap(cm,annot=True)
```

## Support Vector Machine (ebf)

```python
rbf_classifier=SVC(kernel="rbf",gamma="auto")
rbf_classifier.fit(X_train,Y_train)

rbf_accuracies=cross_val_score(estimator=rbf_classifier,X=X_train,y=Y_train,cv=12)
print("Accuracies:\n",rbf_accuracies)
print("Mean Accuracy:",rbf_accuracies.mean())
```

```python
print("Accuracy of RBF model")
rbf_pred=rbf_classifier.predict(X_test)
accuracy_score(Y_test,rbf_pred)
```

```python
rbf_pred=rbf_classifier.predict(X_test)
confusion_matrix(Y_test,rbf_pred)
```

```python
print("Confusion metrix of rbf")
cm=confusion_matrix(Y_test,rbf_pred,labels=[0,1])
sns.heatmap(cm,annot=True)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc_classifier =RandomForestClassifier(n_estimators=40,criterion='entropy',random_state=0)
rfc_classifier.fit(X_train,Y_train)

rfc_accuracies=cross_val_score(estimator=rfc_classifier,X=X_train,y=Y_train,cv=12)
print("Accuracies:\n",rfc_accuracies)
print("Mean Accuracy:",rfc_accuracies.mean())
```

```python
print("Confusion matrix of RandomForestClassifier")
rfc_pred=rfc_classifier.predict(X_test)
confusion_matrix(Y_test,rfc_pred)
```

```python
print("Confusion matrix of RandomForestClassifier")
cm=confusion_matrix(Y_test,rfc_pred)
sns.heatmap(cm,annot=True)
```

## KNeighbors Classifiers

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, GridSearchCV
```

```python
param_grid= {
    'n_neighbors': [2,5,7,9,11],
    "leaf_size" : [2,5,7,9,11],
    "p" :[1,2]   
}
grid = GridSearchCV(KNeighborsClassifier(),param_grid=param_grid)
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
KNN_model = KNeighborsClassifier(leaf_size= 2, n_neighbors= 2, p= 1)
```

```python
KNN_pred = KNN_model.predict(X_test)
accuracy_score(Y_test,KNN_pred)
```

```python
print("Confusion matrix of KNN")
cm=confusion_matrix(Y_test,KNN_pred)
sns.heatmap(cm,annot=True)
```
